```python
#Definisi Node dalam Doubly Linked LIst
class PageNode:
  def __init__(self, folder):
    self.folder = folder
    self.next = None  #Pointer ke folder berikutnya
    self.prev = None  #Pointer ke folder sebelumnya

#Definisi Doubly LInked List untuk Riwayat Folder
class FolderHistory:
  def __init__(self):
    self.current = None  #Pointer ke folder yang sedang dibuka

  #Menambahkan folder baru ke dalam riwayat
  def visit_page(self, folder):
    new_page = PageNode(folder)
    if not self.current:
      self.current = new_page  #Jika riwayat kosong, folder ini menjadi folder pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page  #Pindahkan posisi ke folder baru

  #Menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n Riwayat kosong.")
      return
    print("\n Daftar Folder:")
    temp = self.current
    while temp.prev:
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.folder}")
      temp = temp.next
      count += 1

  #Navigasi mundur ke folder sebelumnya
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\n Anda kembali ke: {self.current.folder}")
    else:
      print("\n Tidak bisa kembali, ini adalah folder pertama!")

  #Navigasi maju ke folder berikutnya
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.folder}")
    else:
      print("\n Tidak bisa maju, ini adalah folder terakhir!")

# ====== Program Utama (Main) ====== #

#Membuat objek riwayat file explorer
history = FolderHistory()

#meminta pengguna memasukkan jumlah folder yang dikunjungi
num_pages = int(input("\n Masukkan jumlah folder yang dikunjungi sesuai digit terakhir NPM:"))

#Loop untuk memasukkan folder ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  folder = input(f"Masukkan Folder  ke-{i}: ")
  history.visit_page(folder)

#Menampilkan riwayat yang telah dimasukkan
history.show_history()

#Menu navigasi pengguna
while True:
  command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar. \nMasukkan perintah: ").lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat file explorer.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

```python
class JadwalNode:
    def __init__(self, mata_kuliah):
        self.mata_kuliah = mata_kuliah
        self.next = None

class JadwalJumat:
    def __init__(self):
        self.head = None

    def tambah_matkul(self, mata_kuliah):
        new_node = JadwalNode(mata_kuliah)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def tampilkan_jadwal(self):
        if not self.head:
            print("Tidak ada mata kuliah!")
            return
        temp = self.head
        count = 1
        print("\nJadwal Mata Kuliah Hari Jumat:")
        while temp:
            print(f"{count}. {temp.mata_kuliah}")
            temp = temp.next
            count += 1

# Program Utama
npm = input("Masukkan digit terakhir NPM Anda: ")
jumlah_matkul = int(npm[-1])  # Ambil digit terakhir

jadwal = JadwalJumat()

for i in range(1, jumlah_matkul + 1):
    matkul = input(f"Masukkan nama mata kuliah ke-{i}: ")
    jadwal.tambah_matkul(matkul)

jadwal.tampilkan_jadwal()
```

```python
class FolderNode:
    def __init__(self, nama_folder):
        self.nama_folder = nama_folder
        self.next = None
        self.prev = None

class FileExplorerHistory:
    def __init__(self):
        self.current = None

    def tambah_folder(self, nama_folder):
        new_folder = FolderNode(nama_folder)
        if not self.current:
            self.current = new_folder
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder

    def tampilkan_riwayat(self):
        if not self.current:
            print("Riwayat kosong!")
            return
        temp = self.current
        while temp.prev:
            temp = temp.prev
        count = 1
        print("\nRiwayat Navigasi Folder:")
        while temp:
            print(f"{count}. {temp.nama_folder}")
            temp = temp.next
            count += 1

# Program Utama
npm = input("Masukkan digit terakhir NPM Anda: ")
jumlah_folder = int(npm[-1])

explorer = FileExplorerHistory()

for i in range(1, jumlah_folder + 1):
    folder = input(f"Masukkan nama folder ke-{i}: ")
    explorer.tambah_folder(folder)

explorer.tampilkan_riwayat()
```

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, data, position):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next and temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        if not temp.next:
            print("Posisi tidak ditemukan!")
            return
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        temp = self.head
        if not temp:
            print("Daftar kosong!")
            return
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# ===== Program Utama (Main) ===== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah elemen awal dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n========== MENU LINKED LIST ==========")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")
    print("=======================================")

    choice = input("\nMasukkan pilihan: ")

    if not choice.isdigit():
        print("Input tidak valid! Masukkan angka.")
        continue

    choice = int(choice)

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(data, position)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

Penambahan Elemen:

Di Awal: Menyisipkan node baru sebagai head.

Di Akhir: Traversal ke node terakhir, lalu hubungkan ke node baru.

Di Posisi Tertentu: Cari node sebelum posisi target, lalu update pointer.

Penghapusan Elemen:

Di Awal: Ubah head ke node berikutnya.

Di Akhir: Traversal ke node sebelum terakhir, lalu set next ke None.

Di Posisi Tertentu: Cari node sebelum target, lalu lewati node target.

Pencarian: Traversal dari head hingga menemukan data yang cocok.

Reverse: Ubah arah pointer setiap node menggunakan variabel bantu prev, current, dan next_node.
